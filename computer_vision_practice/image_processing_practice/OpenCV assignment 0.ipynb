```python
from matplotlib import pyplot as plt
import cv2
import numpy as np
```

```python
Ferrari_img=cv2.imread('E:\\downloads\\ferrari image.jpg')
```

```python
# showing RGB version of the image
cv2.imshow('image',Ferrari_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#rotate by 90 degrees
#ROTATE 90 Counter Clock wise and Rotate 180 also exist
IR=cv2.rotate(Ferrari_img, cv2.ROTATE_90_CLOCKWISE)
```

```python
# showing rotated image
cv2.imshow('Rotated image',IR)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#shape of image #3 due to 3 channels
#(h,width,3 chanels)
Ferrari_img.shape
```

```python
#assigning height and width as dimensions
h=int(Ferrari_img.shape[0]*0.5)#height scaled by 0.5
w=int(Ferrari_img.shape[1]*0.5)#width scaled byh 0.5
dimensions=(w,h)
IResize=cv2.resize(Ferrari_img, dimensions, interpolation = cv2.INTER_AREA)
```

```python
cv2.imshow("resized image by 0.5", IResize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#sobel filter
sobelhorizontal = cv2.Sobel(Ferrari_img,cv2.CV_64F,1,0,ksize=5)#kernels of size 5 
```

```python
#horizontal sobel filter
cv2.imshow("Sobel Filter", sobelhorizontal)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#image enchancement #1 edge detection
edge_Det = cv2.Canny(Ferrari_img,100,200)
cv2.imshow("Edge Detection", edge_Det)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#image enchancement #2 BGR to HSV
New_image = cv2.cvtColor(Ferrari_img,cv2.COLOR_BGR2HSV)
cv2.imshow("BGR to HSV ", New_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#splitting image displaying r
b, g, r = cv2.split(Ferrari_img)
cv2.imshow("red",r)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# splitting image blue

cv2.imshow("blue",b)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#splitting image displaying green
cv2.imshow("green",g)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#converting image to gray scale
grayscale=cv2.imread('E:\\downloads\\ferrari image.jpg',cv2.IMREAD_GRAYSCALE)
cv2.imshow("grayscale image",grayscale)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#thresholding Set threshold T value to 125 Values greater than 125 get assigned 1 the rest are zero
# gray scale image is used for thresholding
r, thresholded_image = cv2.threshold(grayscale, 125, 255, cv2.THRESH_BINARY)
cv2.imshow("thresholded_image",thresholded_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#histogram of original model
plt.hist(grayscale.ravel(),256,[0,256]); plt.show()
```

```python
#display image after histogram equalization Better Contrast the contrast increases
Eimg = cv2.equalizeHist(grayscale)
cv2.imshow("Histogram equalization",Eimg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.hist(Eimg.ravel(),256,[0,256]); 
plt.show()
```
